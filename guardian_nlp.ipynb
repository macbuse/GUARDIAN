```python
import pandas as pd

df = pd.read_csv('transgender.csv')
```

```python
urls = df[df['Tag'] == 'lifeandstyle']['URL'].values
```

```python
from goose3 import Goose
import time, os

def miner(urls, 
          start=0, 
          prefix=''):
    
    #with open('urls.txt','r') as fp:
    #    urls = list(fp.readlines())

    gg = Goose()
    failed = []
    for k, url in enumerate(urls[start:]):
        
        url = url.strip()
        print('Getting', url)
        ss = str(k + start)
        fn = prefix + '0'*(4 - len(ss)) + ss
        try:
            art = gg.extract( url=url)
            with open(fn,'w') as fp:    
                fp.write('\n\n'.join([art.title, art.meta_keywords, art.cleaned_text, url]))
        except:
            failed.append( url)
        time.sleep(5)

    with open('failed.txt','w') as ffp:
        ffp.write('\n'.join(failed))

#this is so that it works nicely on a restart    
prefix, start  = 'll', 0
fns = [ int(x[len(prefix):]) for x in os.listdir('.') if x[:len(prefix)] ==  prefix ]
if fns : start = max(fns) + 1
    
print(start)
miner(urls, start=start, prefix=prefix)
    
```

```python

from spacy.matcher import Matcher
nlp = spacy.load("en_core_web_sm")

```

```python
matcher = Matcher(nlp.vocab)
pattern = [{"LOWER" : "gender"},
           {"POS": "NOUN"}
          ]
matcher.add("GENDER", None, pattern)
```

```python
import re

for k in range(50,100):

    with open('ll{:04}'.format(k),'r') as fp:
        txt = fp.read()
    
    txt = '\n'.join(txt.split('\n'*2)[2:-1]) 
    txt = re.sub(r'”,', '”',txt)
    doc = nlp(txt)
    

    matches = matcher(doc)
    for match_id, start, end in matches:
        span = doc[start:end]
        sent = span.sent 
        print(k, sent.text.strip())
```
